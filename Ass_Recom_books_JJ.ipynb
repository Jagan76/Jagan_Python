```python
# Apply Cosine Logic for Receommendation
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(book_rating_round.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.head()
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books.title.unique()
user_sim_df.columns = books.title.unique()
```

```python
user_sim_df.head()
```

```python
# fill the column value as 0 if userid matches the Column name - Meaning diagonal
np.fill_diagonal(user_sim, 0)
```

```python
# Apply Correlation Logic for Recommendation
```

```python
## Pivot table
book1=book_rating_round.pivot_table(index='Frequency',columns='title', values='Average Rating')
```

```python
book1
```

```python
#Impute those NaNs with 0 values
book1.fillna(0, inplace=True)
```

```python
book1
```

```python
# Stardust is the best rated book  which have the highest rating  among the books rated by 2 or more people. 
# taking Star dust as example for correlation
```

```python
book_star=book1['Stardust']
```

```python
book_star
```

```python
similar_book=book1.corrwith(book_star)
```

```python
similar_book.head()
```

```python
similar_book.head(20)
```

```python
similar_book_df=pd.DataFrame(similar_book,columns=['Correlation'])
```

```python
similar_book_df
```

```python
similar_book_df.sort_values('Correlation',ascending=False).head(10)
```
