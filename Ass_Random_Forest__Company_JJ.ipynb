```python
# Random Forest Classification
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
company = pd.read_csv('Company_Data.csv')
```

```python
company
```

```python
company.info()
```

```python
len(company)
```

```python
company['Sales_New']='S'
company
```

```python
for i in range (0,len(company)):
    if company.Sales[i] < company.Sales.mean():
        company.Sales_New[i]=0   # 0 Indicates low_sales 
    else:
        company.Sales_New[i]=1   # 1 Indicates High_sales
```

```python
company.head()
```

```python
company.Sales_New.unique()
```

```python
company.Sales_New.value_counts()
```

```python
#Dropping "Sales" column
company=company.drop(['Sales'],axis=1)
```

```python
company.head()
```

```python
#Apply Label encoding on categorical data
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
company['ShelveLoc']= label_encoder.fit_transform(company['ShelveLoc'])
company['Urban']= label_encoder.fit_transform(company['Urban']) 
company['US']= label_encoder.fit_transform(company['US']) 
```

```python
company.head()
```

```python
X=company.iloc[:,:-1]
```

```python
Y=company.iloc[:,-1:]
```

```python
X
```

```python
Y
```

```python
# Convert X and Y into Array
import numpy as np
```

```python
x=np.asarray(X,dtype=np.int)
x
```

```python
y=np.asarray(Y,dtype=np.int)
y
```

```python
# Random Forest Classification
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
#Model1
num_trees = 100
max_features = 3
kfold = KFold(n_splits=10, random_state=7)
model1 = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model1, x, y, cv=kfold)
print(results.mean())
```
