```python
idx = 0
user_id = books.index  # can be a list, a Series, an array or a scalar   
books.insert(loc=idx, column='user_id', value=user_id)
```

```python
books
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( books.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books.user_id.unique()
user_sim_df.columns = books.user_id.unique()
```

```python
user_sim_df.head()
```

```python
# fill the column value as 0 if userid matches the Column name - Meaning diagonal
np.fill_diagonal(user_sim, 0)
```

```python
user_sim_df.head()
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:30]
```
